```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("/kaggle/input/rank-salary/rank_salary.csv")
df.head()
```

```python
df= df.drop(columns = ['Position'])
x_data = df['Level']
y_data = df['Salary']
```

```python
slope, intercept, r, p, std_err = stats.linregress(x_data, y_data)
print("the correlation coefficient:" ,r)
```

```python
mymodel = np.poly1d(np.polyfit(x_data,y_data,3))
```

```python
myline = np.linspace(1,10,1000000)
```

```python
plt.scatter(x_data,y_data)
plt.plot(myline,mymodel(myline))
plt.show()
```
