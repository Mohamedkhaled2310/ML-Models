```python
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv('/kaggle/input/multiple-linear-regression-dataset/multiple_linear_regression_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
x_data = df[['age','experience']]
y_data =df['income']
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_data,y_data)
```

```python
X_new = pd.DataFrame([[43, 14]])
predictedIncome = reg.predict(X_new)
```
